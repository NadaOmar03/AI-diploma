```python
a = "Extracurricular Activities"

for no , i in enumerate(df.columns):
    if i != a :
        sns.histplot(data=df,x=i)
        plt.show()
```

```python
for i in ['Hours Studied', 'Sleep Hours','Sample Question Papers Practiced']:
    sns.countplot(x=i,hue="Extracurricular Activities",data=df)
    plt.legend(bbox_to_anchor = (1.05,1))
    plt.show()
```

```python
plt.scatter(df["Hours Studied"],df["Performance Index"])


plt.xlabel("Hours studied")
plt.ylabel("Performance Index")
plt.title("Hours Studied VS Performance Index")
plt.show()
```

```python
sns.barplot(x="Extracurricular Activities", y="Performance Index",data=df)

plt.xlabel("Extracurricular Activities")
plt.ylabel("Performance Index")
plt.title("Performance Index by Extracurricular Activities")
plt.show()
```

```python
%matplotlib inline
plt.xlabel('Extracurricular Activities')
plt.ylabel('')
plt.scatter(df['Extracurricular Activities'],color='red',marker='+')
```

```python
%matplotlib inline
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.scatter(df['Hours Studied'],df['Performance Index'],color='red',marker='.')
```

```python
# Group the data by extracurricular activities and calculate the mean performance
grouped_data = df.groupby('Extracurricular Activities')['Performance Index'].mean()

# Create a bar plot
plt.bar(x=grouped_data.index, y=grouped_data.value)
plt.xlabel('Extracurricular Activities')
plt.ylabel('Mean Performance Index')
plt.title('Impact of Extracurricular Activities on Performance')
plt.show()
```
