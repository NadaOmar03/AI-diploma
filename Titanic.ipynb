```python
df = df.fillna(float('nan'))
df
```

```python
x = df.drop('Survived',axis='columns')
y = df.Survived
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
le_Name= LabelEncoder()
le_Ticket = LabelEncoder()
le_Cabin = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
x['Sex_n'] = le_Sex.fit_transform(x['Sex'])
x['Name_n'] = le_Name.fit_transform(x['Name'])
x['Ticket_n'] = le_Ticket.fit_transform(x['Ticket'])
x['Cabin_n'] = le_Cabin.fit_transform(x['Cabin'])
x['Embarked_n'] = le_Embarked.fit_transform(x['Embarked'])

x_n
```

```python
x_n = x.drop(['Sex', 'Name', 'Cabin', 'Embarked' , 'Ticket'],axis='columns')
x_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_n, y)
```

```python
model.score(x_n,y)
```
