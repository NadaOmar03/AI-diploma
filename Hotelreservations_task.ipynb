```python
import pandas as pd
```

```python
df=pd.read_csv('/Hotel Reservations.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.	Booking_ID=le.fit_transform(df.Booking_ID)
df.type_of_meal_plan=le.fit_transform(df.type_of_meal_plan)
df.	room_type_reserved=le.fit_transform(df.room_type_reserved)
df.market_segment_type=le.fit_transform(df.market_segment_type)


df
```

```python
target= df.booking_status
input= df.drop('booking_status', axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input,target,test_size=0.9,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
